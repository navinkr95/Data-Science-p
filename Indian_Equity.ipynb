```python
# Above we can see that date is a object type ,so i have to change the date to date-datatype
df[' DATE1'] = pd.to_datetime(df[' DATE1'])
df['Month'] = df[' DATE1'].dt.month
df['Year'] = df[' DATE1'].dt.year
```

```python
df = df.sort_values(by=' DATE1')
```

```python
df.info()
```

```python
# to check null values
df.isnull().sum()
```

```python
# removing that row who have null values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# calculate the percentage gain or loss based on prev close with today's close.
df['%_change'] = ( ( df[' CLOSE_PRICE'] - df[' PREV_CLOSE'] )/df[' PREV_CLOSE'] )*100
```

```python
data = df.groupby(['SYMBOL','Year','Month']).agg({'%_change': ['sum']})
data
```

```python
data_group_new = data.reset_index()
```

```python
# Now go through every year and every month to find top 10  best performing stocks.
count = 0
for i in data_group_new['Year'].unique():
    
    new = data_group_new[data_group_new['Year'] == i]
    for j in new['Month'].unique():
        
        
        listofindex = new[new['Month'] == j]['%_change'].sort_values(by='sum',ascending=False).head(10).index.to_list()
        count += 1
        print("IN year {i}: IN {j} Month : Top 10 Performing Stocks :".format(i=i,j=j))
        
        
        for k in listofindex:
            print(df['SYMBOL'].iloc[k])
        print('\n')
print("Total Months:",count)
```

```python
# Now go through every year and every month to find top 10 non best performing stocks.
count = 0
for i in data_group_new['Year'].unique():
    
    new = data_group_new[data_group_new['Year'] == i]
    for j in new['Month'].unique():
        listofindex = new[new['Month'] == j]['%_change'].sort_values(by='sum',ascending=False).tail(10).index.to_list()
        count += 1
        print("IN year {i}: IN {j} Month : Top 10 Non Performing Stocks :".format(i=i,j=j))
        for k in listofindex:
            print(df['SYMBOL'].iloc[k])
        print('\n')
print("Total Months:",count)
```
